```python
# Load dataset
df = pd.read_csv("adult.csv")


# View first rows
print(df.head())
print(df.info())
```

```python
# Replace ? with NaN
df.replace("?", np.nan, inplace=True)

# Check missing values
df.isnull().sum()
```

```python
missing_cols = ['workclass', 'occupation', 'native-country']

for col in missing_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```
